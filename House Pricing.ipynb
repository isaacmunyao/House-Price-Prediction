```python
cat = train.select_dtypes(include='object')
cat
```

```python
#label encoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for c in cat:
    cat[c] = le.fit_transform(cat[c])
```

```python
cat.dtypes
```

```python
numeric = train.select_dtypes(exclude='object')
numeric
```

```python
new_df = pd.concat([cat, numeric],axis=1)
new_df.info()
```

```python
y=new_df['SalePrice']
x = new_df.drop('SalePrice', axis=1)
```

```python
#Standardize the features
# Import the standard scalar
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_std = scaler.fit_transform(x)
```

```python
#Split the data
# Import train test split
from sklearn.model_selection import GridSearchCV, cross_val_score, cross_val_predict, train_test_split

# Using sklearn to split the data into train and test
x_train, x_test, y_train, y_test = train_test_split(x_std, y, test_size=0.3, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Fitting Linear regression
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
predictions = lr.predict(x_test)
```

```python
# Use score method to get accuracy of model
score = lr.score(x_test, y_test)
print(score*100)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor
model_randomforest = RandomForestRegressor(n_estimators=500, n_jobs=-1, random_state=13)
model_randomforest.fit(x_train, y_train)
```

```python
#Scoring model
score_test=model_randomforest.score(x_test, y_test)*100
```
